```python
from langchain_community.tools import DuckDuckGoSearchResults
search=DuckDuckGoSearchResults()

@tool
def multiply(a:int,b:int)->int:
    """This tool is used to multiply 2 numbers"""
    return a*b

@tool
def add(a:int,b:int)->int:
    """This tool is used to add 2 numbers"""
    return a+b

@tool
def divide(a:int,b:int)->float:
    """This tool is used to divide 2 numbers"""
    return a/b
```

```python
tools=[multiply,add,divide,search,weather_tool]
llm_with_tools=chat_model.bind_tools(tools)
```

```python
template="""
    You are a helpful AI travel planner.

    The user will  say about city name, currency etc(if currency or user's place not mentioned consider user's currency as USD),  
    You have access to these tools provided and have to give information about:

    Your task:
     search attraction and activity:
        1. search attracation
        2. search restaurant
        3. search activity
        4. search transportation
    search weather forcasting:
        1. get weather forcast on trip days
    search hotel costs
        1. search hotel
        2. estimate the hotel cost
        3. budget_range
    calculate total cost
        1. calculated total cost
        2. calcualte the daily budget
    currency_converion
        1. get exchange rate
        2. convert currency
    Itinery generation
        1. get day plan
        2. create full itinery
    create Trip Summary

    Always return clear, structured responses.
    Mention Travel details from their location to Destination if they mention their place, otherwise Leave it.
    Provide direct links to purchase tickets, hotel rooms etc.
    And produce trip deatils in 3 types - cheap,moderate,expensive packages and mention about offers also
"""
```

```python
def Guide(state:MessagesState):
    query=state["messages"]
    statement=[template]+query
    print(statement)
    response=llm_with_tools.invoke(statement)
    return {"messages":[response]}
```

```python
workflow=StateGraph(MessagesState)
workflow.add_node("GUIDE",Guide)
workflow.add_node("tools",ToolNode(tools))

workflow.add_edge(START,"GUIDE")
workflow.add_conditional_edges("GUIDE",
                               tools_condition
                               )
workflow.add_edge("tools","GUIDE")
app=workflow.compile()
```

```python
display(Image(app.get_graph(xray=True).draw_mermaid_png()))
```

```python
query={"messages":["I am living in Singapore and i want to go to Switzerland only me i am 18 on 1st week of december 2025"]}
```

```python
events=app.stream(query,stream_mode="values")
```

```python
for e in events:
    e["messages"][-1].pretty_print()
```

```python
response=app.invoke(query)
```
